```python
import pandas as pd
import pandas_ta as ta
import numpy as np

df = df.sort_values(["Ticker", "Date"])
print(f"Initial shape: {df.shape}")

def create_ga_daily_features_group(df_group):
    # Standardize group object
    group = df_group.copy()
    
    # Basic indicators
    group["EMA50"] = ta.ema(group["Close"], length=50)
    group["EMA200"] = ta.ema(group["Close"], length=200)
    group["EMA_RATIO"] = group["EMA50"] / group["EMA200"]

    # MACD
    macd = ta.macd(group["Close"])
    if macd is not None and not macd.empty:
        group["MACD"] = macd.iloc[:, 0]  # MACD_12_26_9
        group["MACD_SIGNAL"] = macd.iloc[:, 1]  # MACDs_12_26_9
        group["MACD_HIST"] = macd.iloc[:, 2]  # MACDh_12_26_9
    else:
        group[["MACD", "MACD_SIGNAL", "MACD_HIST"]] = np.nan

    group["RSI14"] = ta.rsi(group["Close"], length=14)

    # Bollinger Bands
    bb = ta.bbands(group["Close"], length=20)
    if bb is not None and not bb.empty:
        bb_cols = bb.columns.tolist()
        bb_mid = [col for col in bb_cols if "BBM" in col][0]
        bb_upper = [col for col in bb_cols if "BBU" in col][0]
        bb_lower = [col for col in bb_cols if "BBL" in col][0]
        group["BB_MID"] = bb[bb_mid]
        group["BB_WIDTH"] = bb[bb_upper] - bb[bb_lower]
    else:
        group[["BB_MID", "BB_WIDTH"]] = np.nan

    group["ATR14"] = ta.atr(group["High"], group["Low"], group["Close"], length=14)
    group["VOLATILITY"] = group["Close"].rolling(14).std()
    group["OBV"] = ta.obv(group["Close"], group["Volume"])
    
    stoch = ta.stoch(group["High"], group["Low"], group["Close"], k=14, d=3)
    if stoch is not None and not stoch.empty:
        group["STOCH_K"] = stoch.iloc[:, 0]  # STOCHk_14_3_3
        group["STOCH_D"] = stoch.iloc[:, 1]  # STOCHd_14_3_3
    else:
        group[["STOCH_K", "STOCH_D"]] = np.nan

    # ================= 5-DAY TARGET ================= #
    group["future_close_5d"] = group["Close"].shift(-5)
    group["ret_pct_5d"] = (group["future_close_5d"] - group["Close"]) / group["Close"] * 100
    group["y_5d_class"] = group["ret_pct_5d"].apply(lambda x: 1 if x > 0.20 else 0)

    group = group.drop(columns=["future_close_5d", "ret_pct_5d"])
    return group

# Apply logic group-wise
print("Calculating indicators for each ticker...")
try:
    # Modern pandas with include_groups=False if Ticker is the key
    df = df.groupby("Ticker", group_keys=True).apply(create_ga_daily_features_group, include_groups=False)
except TypeError:
    # Fallback for older pandas
    df = df.groupby("Ticker", group_keys=True).apply(create_ga_daily_features_group)

print(f"Shape after feature gen: {df.shape}")

# Diagnostic check for nulls
nulls = df.isnull().sum()
print(f"Null counts before dropna:\n{nulls[nulls > 0]}")

df = df.dropna().reset_index()
# If reset_index created a duplicate 'Ticker' column (one from index, one from data), drop the duplicate
if 'Ticker' in df.columns and hasattr(df.columns, 'duplicated') and df.columns.duplicated().any():
    df = df.loc[:, ~df.columns.duplicated()]

print(f"Final shape after dropna: {df.shape}")
if not df.empty:
    print(df.head())
else:
    print("WARNING: DataFrame is STILL empty after dropna!")
```

```python
import yfinance as yf
import pandas as pd
import pandas_ta as ta
import numpy as np
from sklearn.preprocessing import StandardScaler
import joblib

xgb_model, scaler = joblib.load("xgb_model_large_cap.pkl")

# ============================== 1. Fetch Data ============================== #
ticker = "TCS.NS"
start_date = "2020-01-01"  # adjust as needed
end_date   = pd.Timestamp.today().strftime("%Y-%m-%d")

df_raw = yf.Ticker(ticker).history(start=start_date, end=end_date)

if df_raw.empty:
    print("No data fetched. Check ticker or date range.")
    exit()

df_raw = df_raw.reset_index()
df_raw.rename(columns={"Date":"Date","Open":"Open","High":"High",
                       "Low":"Low","Close":"Close","Volume":"Volume"}, inplace=True)

# ===================== 2. Feature Engineering ====================== #
def create_features(df):
    df = df.copy()
    
    df["EMA50"] = ta.ema(df["Close"], length=50)
    df["EMA200"] = ta.ema(df["Close"], length=200)
    df["EMA_RATIO"] = df["EMA50"] / df["EMA200"]

    macd = ta.macd(df["Close"])
    df["MACD"] = macd["MACD_12_26_9"]
    df["MACD_SIGNAL"] = macd["MACDs_12_26_9"]
    df["MACD_HIST"] = macd["MACDh_12_26_9"]

    df["RSI14"] = ta.rsi(df["Close"], length=14)

    bb = ta.bbands(df["Close"], length=20)
    bb_cols = bb.columns.tolist()
    bb_mid   = [c for c in bb_cols if "BBM" in c][0]
    bb_upper = [c for c in bb_cols if "BBU" in c][0]
    bb_lower = [c for c in bb_cols if "BBL" in c][0]
    df["BB_MID"] = bb[bb_mid]
    df["BB_WIDTH"] = bb[bb_upper] - bb[bb_lower]

    df["ATR14"] = ta.atr(df["High"], df["Low"], df["Close"], length=14)
    df["VOLATILITY"] = df["Close"].rolling(14).std()

    df["OBV"] = ta.obv(df["Close"], df["Volume"])
    stoch = ta.stoch(df["High"], df["Low"], df["Close"], k=14, d=3)
    df["STOCH_K"] = stoch["STOCHk_14_3_3"]
    df["STOCH_D"] = stoch["STOCHd_14_3_3"]

    df = df.dropna().reset_index(drop=True)
    return df

df_features = create_features(df_raw)

# ===================== 3. Regime ====================== #
df_features["Return20"] = df_features["Close"].pct_change()
df_features["Regime"] = df_features["Return20"].rolling(20).mean().apply(lambda x: "Bullish" if x>0 else "Bearish")

# ===================== 4. Hybrid Strategy ====================== #
feature_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'EMA50', 'EMA200',
                'EMA_RATIO', 'MACD', 'MACD_SIGNAL', 'MACD_HIST', 'RSI14', 
                'BB_MID', 'BB_WIDTH', 'ATR14', 'VOLATILITY', 'OBV', 
                'STOCH_K', 'STOCH_D']

Xt = df_features.iloc[[-1]][feature_cols]  # last row only
Pt = df_features.iloc[-1]["Close"]
current_regime = df_features.iloc[-1]["Regime"]

def hybrid_signal(Xt_row, Pt, Cash, model, scaler, regime):
    X_scaled = scaler.transform(Xt_row)
    y_hat = model.predict(X_scaled)[0]  # 0 = Sell, 1 = Buy

    # Extract indicators
    EMA50  = Xt_row["EMA50"].values[0]
    EMA200 = Xt_row["EMA200"].values[0]
    MACD   = Xt_row["MACD"].values[0]
    MACD_SIGNAL = Xt_row["MACD_SIGNAL"].values[0]
    RSI14  = Xt_row["RSI14"].values[0]
    ATR    = Xt_row["ATR14"].values[0]
    VOL    = Xt_row["VOLATILITY"].values[0]
    STOCH_K = Xt_row["STOCH_K"].values[0]
    STOCH_D = Xt_row["STOCH_D"].values[0]

    # ========================= BUY LOGIC ========================= #
    score_buy = 0

    if Pt > EMA50 and MACD > MACD_SIGNAL:
        score_buy += 1.5
    if RSI14 < 40 and y_hat == 1:
        score_buy += 1.2
    if Pt > EMA200 and regime == "Bullish":
        score_buy += 1
    if STOCH_K < 20 and STOCH_K < STOCH_D:
        score_buy += 1.2

    if score_buy >= 2.5:
        risk_trade = 0.01 * Cash
        shares_at_risk = risk_trade / ATR if ATR != 0 else 0
        shares_cap = 0.1 * Cash / Pt
        shares = min(shares_at_risk, shares_cap)
        return "Buy", int(shares)

    # ========================= SELL LOGIC ========================= #
    if y_hat == 0:
        score_sell = 0

        if RSI14 > 65:
            score_sell += 1.2
        if Pt < EMA200:
            score_sell += 1.0
        if Pt < EMA50:
            score_sell += 0.8
        if MACD < MACD_SIGNAL:
            score_sell += 0.8
        if STOCH_K > 80 and STOCH_K > STOCH_D:
            score_sell += 0.6
        if VOL > 0.03 * Pt:
            score_sell += 0.5

        if score_sell >= 1.5:
            return "Sell", "Full"

    # ========================= DEFAULT RETURN ========================= #
    return "Hold", 0

# ===================== 5. Output ====================== #
signal, position = hybrid_signal(
    Xt, Pt, Cash=100000, model=xgb_model, scaler=scaler, regime=current_regime
)

print(f"Ticker: {ticker}")
print(f"Date: {df_features.iloc[-1]['Date']}")
print(f"Signal: {signal}")
print(f"Shares: {position}")

```

```python
# ===================== IMPORTS =====================
import yfinance as yf
import pandas as pd
import pandas_ta as ta
import numpy as np
from datetime import datetime, timedelta
import joblib

# ===================== LOAD MODEL & SCALER =====================
# Model trained for LARGE CAP stocks
model, scaler = joblib.load("xgb_model_large_cap.pkl")


# ===================== FUNCTION TO CREATE FEATURES =====================
def create_features_for_ticker(ticker_symbol, days=200):
    """
    Download data for a ticker and create all required features
    """

    try:
        # Download historical data
        end_date = datetime.today()
        start_date = end_date - timedelta(days=days)

        print(f"📥 Downloading data for {ticker_symbol}...")
        data = yf.download(
            ticker_symbol,
            start=start_date,
            end=end_date,
            interval="1d",
            progress=False
        )

        if data.empty:
            print(f"❌ No data found for {ticker_symbol}")
            return None

        # Reset index
        data = data.reset_index()

        # Handle MultiIndex columns (yfinance issue)
        if isinstance(data.columns, pd.MultiIndex):
            data.columns = data.columns.droplevel(1)

        print(f"   Downloaded {len(data)} rows of data")

        # ================= CREATE FEATURES ================= #

        # TREND FEATURES
        data['EMA50'] = ta.ema(data['Close'], length=50)
        data['EMA200'] = ta.ema(data['Close'], length=200)

        data['EMA_RATIO'] = np.where(
            data['EMA200'] != 0,
            data['EMA50'] / data['EMA200'],
            1.0
        )

        # MACD
        macd_result = ta.macd(data['Close'], fast=12, slow=26, signal=9)
        if macd_result is not None and not macd_result.empty:
            macd_cols = macd_result.columns.tolist()
            data['MACD'] = macd_result[[c for c in macd_cols if "MACD_" in c and "s" not in c][0]]
            data['MACD_SIGNAL'] = macd_result[[c for c in macd_cols if "MACDs_" in c][0]]
            data['MACD_HIST'] = macd_result[[c for c in macd_cols if "MACDh_" in c][0]]
        else:
            data[['MACD', 'MACD_SIGNAL', 'MACD_HIST']] = 0

        # RSI
        data['RSI14'] = ta.rsi(data['Close'], length=14)

        # BOLLINGER BANDS
        bb_result = ta.bbands(data['Close'], length=20)
        if bb_result is not None and not bb_result.empty:
            bb_cols = bb_result.columns.tolist()
            data['BB_MID'] = bb_result[[c for c in bb_cols if "BBM" in c][0]]
            data['BB_WIDTH'] = (
                bb_result[[c for c in bb_cols if "BBU" in c][0]] -
                bb_result[[c for c in bb_cols if "BBL" in c][0]]
            )
        else:
            data['BB_MID'] = data['Close']
            data['BB_WIDTH'] = 0

        # ATR
        atr_result = ta.atr(data['High'], data['Low'], data['Close'], length=14)
        data['ATR14'] = atr_result if atr_result is not None else (data['High'] - data['Low'])

        # VOLATILITY
        data['VOLATILITY'] = data['Close'].rolling(14).std()

        # OBV
        obv_result = ta.obv(data['Close'], data['Volume'])
        data['OBV'] = obv_result if obv_result is not None else 0

        # STOCHASTIC OSCILLATOR
        stoch_result = ta.stoch(data['High'], data['Low'], data['Close'], k=14, d=3)
        if stoch_result is not None and not stoch_result.empty:
            stoch_cols = stoch_result.columns.tolist()
            data['STOCH_K'] = stoch_result[[c for c in stoch_cols if "STOCHk_" in c][0]]
            data['STOCH_D'] = stoch_result[[c for c in stoch_cols if "STOCHd_" in c][0]]
        else:
            data['STOCH_K'] = 50
            data['STOCH_D'] = 50

        # CLEANUP
        data = data.ffill().bfill().fillna(0)
        data = data.dropna(subset=['Close', 'Volume'])

        print(f"✅ Features created! Total rows: {len(data)}")
        return data

    except Exception as e:
        print(f"❌ Error processing {ticker_symbol}: {str(e)}")
        return None


# ===================== FUNCTION TO MAKE PREDICTIONS =====================
def predict_5d(ticker_symbol, days=200):
    """
    Predict price movement over next 5 days
    """

    data = create_features_for_ticker(ticker_symbol, days)

    if data is None or len(data) < 50:
        print(f"❌ Insufficient data for {ticker_symbol}")
        return None

    feature_columns = [
        'Open', 'High', 'Low', 'Close', 'Volume',
        'EMA50', 'EMA200', 'EMA_RATIO',
        'MACD', 'MACD_SIGNAL', 'MACD_HIST',
        'RSI14', 'BB_MID', 'BB_WIDTH',
        'ATR14', 'VOLATILITY',
        'OBV', 'STOCH_K', 'STOCH_D'
    ]

    latest_data = data.iloc[-1:][feature_columns].copy()

    if latest_data.isnull().any().any():
        latest_data = latest_data.fillna(latest_data.mean())

    # SCALE + PREDICT
    latest_scaled = scaler.transform(latest_data)
    prediction_prob = model.predict_proba(latest_scaled)[0]
    prediction_class = model.predict(latest_scaled)[0]

    current_price = data.iloc[-1]['Close']
    prediction_date = data.iloc[-1]['Date']

    if prediction_class == 1:
        signal = "🟢 UP"
        confidence = prediction_prob[1]
    else:
        signal = "🔴 DOWN"
        confidence = prediction_prob[0]

    return {
        'Ticker': ticker_symbol,
        'Date': prediction_date,
        'Current_Price': current_price,
        'Prediction': signal,
        'Confidence': confidence,
        'UP_Probability': prediction_prob[1],
        'DOWN_Probability': prediction_prob[0]
    }


# ===================== RUN PREDICTION =====================
print("\n" + "=" * 60)
print("SINGLE STOCK PREDICTION (LARGE CAP MODEL)")
print("=" * 60)

result = predict_5d("INFY.NS")

if result:
    print("\n✅ PREDICTION SUCCESSFUL!")
    print(f"Ticker: {result['Ticker']}")
    print(f"Date: {result['Date']}")
    print(f"Current Price: ₹{result['Current_Price']:.2f}")
    print(f"Prediction: {result['Prediction']}")
    print(f"Confidence: {result['Confidence']:.4f} ({result['Confidence']*100:.2f}%)")
    print(f"UP Probability: {result['UP_Probability']:.4f}")
    print(f"DOWN Probability: {result['DOWN_Probability']:.4f}")
else:
    print("❌ Prediction failed.")
```

```python
import yfinance as yf
import pandas as pd
import pandas_ta as ta
import numpy as np
from sklearn.preprocessing import StandardScaler
import joblib

# ===================== 1. Load Model & Scaler ===================== #
model_data = joblib.load("xgb_model_large_cap.pkl")
xgb_model = model_data[0]  # model
scaler    = model_data[1]  # scaler

# ===================== 2. Fetch Midcap Data ===================== #
ticker = "MARUTI.NS"  # replace with your midcap ticker
start_date = "2020-01-01"
end_date = pd.Timestamp.today().strftime("%Y-%m-%d")

df_raw = yf.Ticker(ticker).history(start=start_date, end=end_date)
if df_raw.empty:
    print("No data fetched. Check ticker or date range.")
    exit()

df_raw = df_raw.reset_index()

def create_features(df):
    df = df.copy()
    
    df["EMA50"] = ta.ema(df["Close"], length=50)
    df["EMA200"] = ta.ema(df["Close"], length=200)
    df["EMA_RATIO"] = df["EMA50"] / df["EMA200"]

    macd = ta.macd(df["Close"])
    df["MACD"] = macd["MACD_12_26_9"]
    df["MACD_SIGNAL"] = macd["MACDs_12_26_9"]
    df["MACD_HIST"] = macd["MACDh_12_26_9"]

    df["RSI14"] = ta.rsi(df["Close"], length=14)

    # ===================== Bollinger Bands (dynamic) ===================== #
    bb = ta.bbands(df["Close"], length=20)
    bb_cols = bb.columns.tolist()
    bb_mid   = [c for c in bb_cols if "BBM" in c][0]
    bb_upper = [c for c in bb_cols if "BBU" in c][0]
    bb_lower = [c for c in bb_cols if "BBL" in c][0]
    df["BB_MID"] = bb[bb_mid]
    df["BB_WIDTH"] = bb[bb_upper] - bb[bb_lower]

    df["ATR14"] = ta.atr(df["High"], df["Low"], df["Close"], length=14)
    df["VOLATILITY"] = df["Close"].rolling(14).std()
    df["OBV"] = ta.obv(df["Close"], df["Volume"])

    stoch = ta.stoch(df["High"], df["Low"], df["Close"], k=14, d=3)
    df["STOCH_K"] = stoch["STOCHk_14_3_3"]
    df["STOCH_D"] = stoch["STOCHd_14_3_3"]

    df = df.dropna().reset_index(drop=True)
    return df

df_features = create_features(df_raw)

# ===================== 4. Regime ===================== #
df_features["Return20"] = df_features["Close"].pct_change()
df_features["Regime"] = df_features["Return20"].rolling(20).mean().apply(
    lambda x: "Bullish" if x > 0 else "Bearish"
)

# ===================== 5. Hybrid Signal Function ===================== #
feature_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'EMA50', 'EMA200',
                'EMA_RATIO', 'MACD', 'MACD_SIGNAL', 'MACD_HIST', 'RSI14', 
                'BB_MID', 'BB_WIDTH', 'ATR14', 'VOLATILITY', 'OBV', 
                'STOCH_K', 'STOCH_D']

def hybrid_signal_midcap(Xt_row, Pt, Cash, model, scaler, regime):
    X_scaled = scaler.transform(Xt_row)
    y_hat = model.predict(X_scaled)[0]  # 0 = Sell, 1 = Buy

    EMA50  = Xt_row["EMA50"].values[0]
    EMA200 = Xt_row["EMA200"].values[0]
    MACD   = Xt_row["MACD"].values[0]
    MACD_SIGNAL = Xt_row["MACD_SIGNAL"].values[0]
    RSI14  = Xt_row["RSI14"].values[0]
    ATR    = Xt_row["ATR14"].values[0]
    VOL    = Xt_row["VOLATILITY"].values[0]
    STOCH_K = Xt_row["STOCH_K"].values[0]
    STOCH_D = Xt_row["STOCH_D"].values[0]

    # --------- Buy Logic (rigorous) --------- #
    score_buy = 0
    if Pt > EMA50 and MACD > MACD_SIGNAL: score_buy += 1.5
    if RSI14 < 40 and y_hat == 1: score_buy += 1.2
    if Pt > EMA200 and regime == "Bullish": score_buy += 1
    if STOCH_K < 20 and STOCH_K < STOCH_D: score_buy += 1.2

    if score_buy >= 2.5:
        risk_trade = 0.01 * Cash
        shares_at_risk = risk_trade / ATR
        shares_cap = 0.1 * Cash / Pt
        shares = min(shares_at_risk, shares_cap)
        return "Buy", int(shares)

    # --------- Sell Logic (mild) --------- #
    score_sell = 0
    if y_hat == 0:
        if RSI14 > 65: score_sell += 1.2
        if Pt < EMA200: score_sell += 1.0
        if Pt < EMA50: score_sell += 0.8
        if MACD < MACD_SIGNAL: score_sell += 0.8
        if STOCH_K > 80 and STOCH_K > STOCH_D: score_sell += 0.6
        if VOL > 0.03 * Pt: score_sell += 0.5
        if score_sell >= 1.5:
            return "Sell", "Full"

    # --------- Default Hold --------- #
    return "Hold", 0

# ===================== 6. Predict Latest Row ===================== #
Xt = df_features.iloc[[-1]][feature_cols]
Pt = df_features.iloc[-1]["Close"]
current_regime = df_features.iloc[-1]["Regime"]

signal, position = hybrid_signal_midcap(
    Xt, Pt, Cash=100000, model=xgb_model, scaler=scaler, regime=current_regime
)

# ===================== 7. Output ===================== #
print(f"Ticker: {ticker}")
print(f"Date: {df_features.iloc[-1]['Date']}")
print(f"Signal: {signal}")
print(f"Shares: {position}")
```
